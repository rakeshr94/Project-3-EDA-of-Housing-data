```python
# Multivariate analysis
# Creation of a sample dataframe
np.random.seed(0)
```

```python
df2 = pd.DataFrame({
    'numeric1': np.random.randn(100),
    'numeric2': np.random.randn(100),
    'category1': np.random.choice(['MSSubClass', 'MSZoning', 'Street'], 100),
    'category2': np.random.choice(['Alley', 'LotShape'], 100),
})
```

```python
df2
```

```python
# Analyze by Numeric vs. Numeric
sns.scatterplot(x='numeric1', y='numeric2', data=df2)
```

```python
# Analyze Numeric vs. Categorical
sns.boxplot(x='category1', y='numeric1', data=df2)
```

```python
# Analyze Categorical vs. Categorical
sns.countplot(x='category1', hue='category2', data=df2)
```

```python
df3 = pd.DataFrame({
    'feature1': np.random.randn(100),
    'feature2': np.random.randn(100),
    'target': np.random.choice([0, 1], 100),
})
```

```python
df3
```

```python
# Analyze Feature vs. Feature
sns.scatterplot(x='feature1', y='feature2', data=df3)
```

```python
# Analyze Feature vs. Target
sns.boxplot(x='target', y='feature1', data=df3)
```

```python
# Ploting the data through Box plot
sns.boxplot(x='LandSlope', y='LotFrontage', data=df1)
```

```python
# Plotting the data through Scatter plot
sns.scatterplot(x='LotFrontage', y='LotArea', data=df1)
```

```python
# Plot the data through Violin plot
sns.violinplot(x='LandSlope', y='LotFrontage', data=df1)
```

```python
# Plot the data through Bubble plot
sns.scatterplot(x='LotFrontage', y='LotArea', size='SalePrice', data=df1)
```

```python
# Plot the data through Histogram
sns.histplot(df1['LotArea'])
```

```python
# Plot the data through Bar plot
sns.barplot(x='LandSlope', y='LotFrontage', data=df1)
```

```python
# Plot the data through Kernel Density Estimation
sns.kdeplot(df1['LotFrontage'])
```

```python
# Plot the data through Correlation plot
sns.heatmap(df1.corr())
```
